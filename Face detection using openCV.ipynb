```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
face_classifier = cv2.CascadeClassifier('haarcascades//haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('new-museum-wafaces-banner-v2.jpg')
plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
def detect_faces(fix_img):
    face_rect = face_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w, y+h),
                     (255, 0, 0),
                     2)
    return fix_img
```

```python
faces = face_classifier.detectMultiScale(img, 1.3, 3)

if faces is ():
    print('no faces found')
else:
    result = detect_faces(fix_img)
    plt.imshow(result)
```
